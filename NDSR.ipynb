# Find negative values, note them, and make positive

```python
df['Absolute value Gram Amount of Food']=abs(df['Gram Amount of Food'])
```

```python
df['Absolute value Gram Amount of Food'].describe()
```

```python
df.shape
```

# Get the frequencies

```python
freq=pd.DataFrame(df['Food Name'].value_counts()).astype(int)
```

```python
freq.columns = ['frequency']
```

# Group by the food names

```python
gf=df.groupby(['Food Name']).sum()
```

# Merge the dataframes

```python
total_df = gf.merge(freq, how='outer', left_index=True, right_index=True)
```

# Look at what the most frequent and the most (volume) consumed foods are

```python
total_df.sort_values(by=['frequency','Absolute value Gram Amount of Food'])
```
